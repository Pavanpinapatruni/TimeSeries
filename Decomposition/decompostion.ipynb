```python
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.seasonal import seasonal_decompose
```

```python
df = pd.read_csv('AirPassengers.csv', header=0, index_col=0, parse_dates=True)
```

```python
## perform seasonal decomposition
result = seasonal_decompose(df, model='multiplicative')
```

```python
# plot the decomposition
plt.rcParams.update({'figure.figsize': (10,10)})
result.plot()
plt.show()
```

```python
result2 = seasonal_decompose(df, model='additive')

plt.rcParams.update({'figure.figsize': (10,10)})
result2.plot()
plt.show()
```

```python
# Add this cell to help decide between models

# Calculate coefficient of variation for different periods
import numpy as np

# Group by month to see seasonal variation
df_monthly = df.groupby(df.index.month)
cv_values = []

for month, group in df_monthly:
    cv = np.std(group.values) / np.mean(group.values)
    cv_values.append(cv)

avg_cv = np.mean(cv_values)
print(f"Average Coefficient of Variation: {avg_cv:.4f}")

if avg_cv > 0.1:
    print("Recommendation: Use MULTIPLICATIVE decomposition")
    print("Seasonal variation changes with trend level")
else:
    print("Recommendation: Use ADDITIVE decomposition") 
    print("Seasonal variation is relatively constant")
```
